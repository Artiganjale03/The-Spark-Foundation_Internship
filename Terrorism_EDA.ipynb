**Renaming Features**

```python
df.columns.values
```

```python
df.rename(columns={
    'iyear': 'year',
    'imonth': 'month',
    'iday': 'day',
    'country_txt': 'Country',
    'region_txt': 'Region',
    'provstate': 'State',
    'gname': 'group',
    'city':'city',
    'latitude':"latitude",
    'longitude':"longitude",
    'summary':'summary',

    'attacktype1_txt': 'attack_type',
    'weaptype1_txt': 'weapon_type',
    'targtype1_txt': 'target_type',
    'nkill': 'kill',
    'nwound': 'wound',

}, inplace=True)
```

```python
df=df[["year",'month','day','Country','Region','city','latitude', 'longitude','summary','success', 'suicide','attack_type','target_type','target1','group','motive','weapon_type','kill','wound']]
```

```python
df
```

**Hndling missing values**

```python
df.isnull().sum()
```

```python
df['wound']=df['wound'].fillna(0)
df['kill']=df['kill'].fillna(0)
```

**Adding one new column:casualities**

```python
df['casualities']=df['kill']+df['wound']
```

```python
# Create a countplot for year-wise terror attacks
plt.figure(figsize=(12, 6))
sns.countplot(x='year', data=df, palette='viridis')
plt.title('Year-wise Terrorist Attacks Count')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

# Plotting the area plot
pd.crosstab(df.year,df.Region).plot(kind='area', stacked=True, figsize=(12, 6))
plt.title('Terrorist Activities by Region Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

**Top 15 countries suffers a lot**

```python
# Group by country and count the number of attacks
attacks_per_country = df['Country'].value_counts()

# Select the top 15 countries
top_15_countries = attacks_per_country.head(15)

# Plotting the bar chart
top_15_countries.plot(kind='bar', figsize=(12, 6))
plt.title('Top 15 Countries Suffering the Most from Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.show()
```

**Year and Region wise casualties**

```python
plt.subplots(figsize=(15,5))
df.groupby(['Region','year']).count()['casualities'].plot()
plt.ylabel('casualities')
plt.show()
```

**Number of attacks by region**

```python
# Count the number of attacks by region
attacks_by_region = df['Region'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=attacks_by_region.index, y=attacks_by_region.values, palette='viridis')
plt.title('Number of Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
weapon_data=df['weapon_type'].value_counts()
weapon_data
```

**Number of attacks by weapons**

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=weapon_data.index, y=weapon_data.values, palette='viridis')
plt.title('Number of Attacks by Weapon Type')
plt.xlabel('Weapon Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Count the occurrences of each city
attacks_by_city = df['city'].value_counts()

# Select the top 10 cities
top_10_cities = attacks_by_city.head(10)

# Plotting the bar chart for the top 10 cities
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_cities.index, y=top_10_cities.values, palette='viridis')
plt.title('Top 10 Cities by Number of Attacks')
plt.xlabel('City')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Count the occurrences of each attack type
attacks_by_type = df['attack_type'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=attacks_by_type.index, y=attacks_by_type.values, palette='viridis')
plt.title('Number of Attacks by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Group by attack type and sum the number of kills
kills_by_attack_type = df.groupby('attack_type')['kill'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='attack_type', y='kill', data=kills_by_attack_type, palette='viridis')
plt.title('Number of Kills by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of Kills')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Exclude rows where 'group' is 'unknown'
df_filtered = df[df['group'] != 'unknown']

# Count the occurrences of each terrorist group
terrorists_by_group = df_filtered['group'].value_counts()

# Select the top 10 terrorist groups
top_10_terrorists = terrorists_by_group.head(10)

# Plotting the bar chart for the top 10 terrorist groups
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_terrorists.index, y=top_10_terrorists.values, palette='viridis')
plt.title('Top 10 Terrorist Groups (Excluding Unknown)')
plt.xlabel('Terrorist Group')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
target_data=df['target_type'].value_counts().nlargest(n=15)
target_data
```

```python
# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=target_data.index, y=target_data.values, palette='viridis')
plt.title('Top 15 Target Types')
plt.xlabel('Target Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Group by region and sum the number of wounded
wounded_by_region = df.groupby('Region')['wound'].sum()

# Select the top 10 regions with the largest number of wounded
top_10_regions = wounded_by_region.nlargest(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_regions.index, y=top_10_regions.values, palette='viridis')
plt.title('Top 10 Regions by Number of Wounded')
plt.xlabel('Region')
plt.ylabel('Number of Wounded')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Group by region and sum the number of killed
killed_by_region = df.groupby('Region')['kill'].sum()

# Select the top 10 regions with the largest number of killed
top_10_regions_killed = killed_by_region.nlargest(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_regions_killed.index, y=top_10_regions_killed.values, color='navy')
plt.title('Top 10 Regions by Number of Killed')
plt.xlabel('Region')
plt.ylabel('Number of Killed')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Count the occurrences of success and failure
success_counts = df['success'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(success_counts, labels=success_counts.index, autopct='%1.1f%%', colors=['navy', 'orange'])
plt.title('Success/Failure Distribution in Terrorist Attacks')
plt.show()
```

```python
# Group by region and sum the number of suicide attacks
suicide_by_region = df.groupby('Region')['suicide'].sum().reset_index()
suicide_by_region
```
