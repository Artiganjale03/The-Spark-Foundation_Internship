**Analysis of the sales**

```python
# Define a function for sales analysis and plotting
def analyze_and_plot_sales(sales_data):
    total_sales = sales_data.sum()
    average_sales = sales_data.mean()
    max_sales = sales_data.max()

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(['Total Sales', 'Average Sales', 'Maximum Sales'], [total_sales, average_sales, max_sales], color=['blue', 'green', 'red'])
    plt.title('Sales Analysis')
    plt.ylabel('Sales')
    plt.show()

    return total_sales, average_sales, max_sales

# Extract the 'Sales' column from the DataFrame
sales_data = df['Sales']

# Call the function to perform the analysis and plot the results
total_sales, average_sales, max_sales = analyze_and_plot_sales(sales_data)

# Display the results
print(f'Total Sales: ${total_sales:.2f}')
print(f'Average Sales: ${average_sales:.2f}')
print(f'Maximum Sales: ${max_sales:.2f}')
```
